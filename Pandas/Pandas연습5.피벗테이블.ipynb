# 1. 피벗테이블

- 가지고 있는 데이터원본을 원하는 형태의 가공된 정보를 보여줌
    - 자료의 형태를 변경하기 위해 많이 사용함 
- pivot_table(data,values=None,index=None, columns=None, aggfunc='mean', margins=False, margins_name='All')
    - data : 분석할 데이터 프레임, 메서드 형식일때는 필요하지 않음 예) df1.pivot_table() 
    - values : 분석할 데이터 프레임에서 분서할 열 
    - index : 행 인덱스로 들어갈 키열 또는 키열의 리스트
    - fill_value : NaN이 표출될 때 대체값 지정
    - margins : 모든 데이터를 분석한 결과를 행으로 표출할지 여부 
    - margins_name : margins가 표출될 때 그열(행)의 이름

**피봇테이블을 작성할 때 반드시 설정해야 되는 인수**

    - data : 사용 데이터 프레임
    - index : 행 인덱스로 사용할 필드(기준 필드로 작용됨)
    - 인덱스 명을 제외한 나머지 값(data)은 수치 data만 사용함
    - 기본 함수가 평균(mean)함수 이기 때문에 각 데이터의 평균값이 반환

# 2. 그룹분석
- 만약 키가 지정하는 조건에 맞는 데이터가 하나 이상이라서 데이터 그룹을 이루는 경우에는 그룹의 특성을 보여주는 그룹분석(group analysis)를 해야 함
    - 그룹 분석은 피봇테이블과 달리 키에 의해서 결정되는 데이터가 여러개가 있을 경우 미리 지정한 연산을 통해 그 그룹 데이터을 계산하는 것

- 판다스에서는 groupby 메서드를 사용하여 아래 내용 처럼 그룹분석을 진행
    - 분석하고자하는 시리즈나 df에 groupby메서드를 호출하여 그룹화 수행
    - 그룹 객체에 대해 그룹 연산을 수행

## (1) groupby 메서드

- 데이터를 그룹별로 분류하는 역할을 함
- 인수
    - 열 또는 열의 리스트
    - 행 인덱스
- 연산 결과로 그룹 데이터를 나타내는 Groupby클래스를 객체로 반환
    - 이 객체에는 그룹별로 연산을 할 수 있는 그룹연산 메서드가 있음


## (2) Groupby 클래스 객체의 그룹연산 메서드
    1. size, count : 그룹 데이터의 갯수
    2. mean, median, min, max : 그룹 데이터의 평균, 중앙값, 최소, 최대값
    3. sum, prod, std, var, quantile :그룹 데이터의 합계, 곱, 표준편차, 분산, 사분위수
    4. first, last : 그룹 데이터 중 가장 첫번째 데이터와 가장 나중 데이터

## (3) 이외에도 많이 사용되는 그룹연산

- agg, aggregate
    - 만약 원하는 그룹연산이 없는 경우 함수를 만들고 이 함수를 agg에 전달
    - 또는 여러가지 그룹연산을 동시에 하고 싶은 경우 함수 이름 문자열의 리스트를 전달
- describe
    - 하나의 그룹 대표값이 아니라 여러개의 값을 데이터프레임으로 구현
- apply
    - describe 처럼 하나의 대표값이 아닌 df을 출력하지만 원하는 그룹연산이 없는 경우에 사용
- transform
    - 그룹에 대한 대표값을 만드는 것이 아니라 그룹별 계산을 통해 데이터자체를 변형

```python
np.random.seed(0)
df2=pd.DataFrame({
    'key1':['A','A','B','B','A'],
    'key2':['one','two','one','two','one'],
    'data1':[1,2,3,4,5],
    'data2':[10,20,30,40,50]
})
df2
```

```python
groups = df2.groupby(df2.key1)
groups
```

```python
pd.DataFrame(groups)
pd.DataFrame(groups).loc[0].values # 0번째 행에 대한 값정보
pd.DataFrame(groups).loc[0].values[1] 
```

```python
#groupby 객체 연산 메서드를 적용시켜 요약 결과를 확인
groups.sum()
```

```python
groups['data1'].sum() # data1 선택후 sum
groups.sum()['data1'] #sum하고 나서 data1 선택 (위와 결과는 같음 )
```

```python
groups[['data1']].sum()
groups[['data2']].sum()

# 참고) 대괄호 두번 : DF반환, 대괄호 한번 : Series반환
```

# 그룹함수 예제
- apply()
- agg()
    - DF 등에 벡터라이징 연산을 적용하는 함수(axis=0/1이용하여 행/열 적용가능)
    - agg함수는 숫자 타입의 스칼라만 리턴하는 함수를 적용

```python
iris=sns.load_dataset("iris")
iris
```

```python
# iris 품종별로 그룹화
i_groups = iris.groupby(iris.species)
i_groups.mean()
```

```python
i_groups.petal_length.sum()
```

```python
i_groups.max()/i_groups.min()
```

```python
def peak_to_peak_ratio(x): #x는 시리즈
    return x.max() / x.min()

peak_to_peak_ratio(iris.sepal_length)
```

```python
# 그룹 객체의 각 그룹에 대해 열별로 peak_to_peak_ratio 함수 호출
i_groups.agg(peak_to_peak_ratio)
i_groups.apply(peak_to_peak_ratio)
```

```python
def top3_petal_length(df):
    return df.sort_values(by="petal_length", ascending=False)[:3]

# 함수 반환값이 수치 집합(df가 반환됨)
top3_petal_length(iris)
```

```python
i_groups.apply(top3_petal_length)
```

## apply예제2

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=['소','중','대']).astype(str)
```

```python
iris.groupby(iris.species).petal_length.apply(q3cut)
```

```python
q3cut(iris.petal_length)
```

```python
iris['petal_length_class']=iris.groupby(iris.species).petal_length.apply(q3cut)
iris.head()
```

# 그룹함수 및 피벗테이블을 이용한 분석 예제
- 식사 후 팁(tip)과 관련된 데이터 셋 사용
    - total_bill : 식사대금
    - tip : 팁
    - sex : 성별
    - smoker : 흡연/금연 여부
    - day : 요일
    - time : 시간
    - size : 인원

```python
tips = sns.load_dataset("tips")
tips.tail()
```

#### 식사 대금 별 팁의 비율이 언제 가장 높은가?
- 가공 필드 생성: 식사대금 대비 팁의 비율
    - tip_pt=팁/식사대금 

```python
tips['tip_pt']=tips['tip']/tips['total_bill']
tips.head()
```

```python
tips.describe()
```

```python
test_df=pd.DataFrame(tips.groupby('sex'))
test_df.loc[1][1].head(2)
```

```python
# 성별 인원수를 계산
# 모든 열에 결측치가 없으므로 모든 열의 성별 인원수는 동일

tips.groupby('sex').count()
tips.groupby('sex').size() # 그룹내의 원소 개수를 반환
```

```python
# 흡연 유무에 따른 성별 인원
t_gp=tips.groupby(['sex','smoker'])
t_gp.size() #남성 여성 모두 비흡연자 수가 많음.
```

```python
# 흡연 유무에 따른 성별 인원
t_gp=tips.groupby(['sex','smoker'])
t_gp.size()

# 남성 여성 모두 비흡연자수가 많음 
```

```python
# 흡연 유무에 따른 성별 인원을 피봇테이블로 구현
tips.pivot_table('total_bill',index='sex',columns='smoker',aggfunc='count')
```

```python
# 성별 팁 비율의 평균
tips.groupby('sex')['tip_pt'].mean()

# 여성이 식사금액 대비 팁 비율 평균이 근소하게 높음
```

```python
# 흡연 유무에 따른 팁 비율의 평균
tips.groupby('smoker')['tip_pt'].mean()

# 흡연자가 비 흡연자에 비해 팁비율이 근소하게 높음 
```

```python
# 성병과 흡연 유무에 따른 팁 비율의 평균
tips.pivot_table(values='tip_pt', index='sex', columns='smoker')

# 결과) 여성 혹은 흡연자의 팁비율이 조금 높고, 여성 흡연자가 팁을 많이 줌
```

```python
# 요일별 손님 수
tips.groupby('day').size()
```

```python
# 요일별 여성/남성 손님 비율
tips.groupby(['day','sex']).size()/len(tips)*100

# 여성 손님은 목요일, 남성 손님은 주말에 비중이 높음
```

```python
# 요일별 테이블당 평균 인원 수 
tips.groupby('day')['size'].mean()
```

```python
# 요일별 평균 팁 비율
tips.groupby('day')['tip_pt'].mean()
```

```python
# 요일별 평균 식사 금액과 평균 팁
tips.groupby('day')[['total_bill','tip']].mean()
```
