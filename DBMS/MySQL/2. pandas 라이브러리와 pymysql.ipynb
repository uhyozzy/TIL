## **2.2 외래키(FOREIGN KEY)**
- 외래키 생성 목적 : 두 테이블 사이에 관계를 선언하여 데이터의 무결성 보장

```python
host_name = 'localhost'
host_port = 3306
username = 'root'
password = 'toor'
database_name = 'sqlDB'

# db연결
db = pymysql.connect(
    host = host_name,      # MySQL Server Address
    port = host_port,      # MySQL Server Port
    user = username,       # password for MySQL username
    passwd = password,     # Database name
    db=database_name,
    charset='utf8'
)
```

```python
sql = "select * from usertbl"
df = pd.read_sql(sql, db)
df
```

```python
sql = "select * from buytbl"
df = pd.read_sql(sql, db)
df
```

**buytbl에 데이터를 추가** (에러가 나야 정상!!)
- CONSTRAINT buytbl_ibfk_1 FOREIGN KEY (user ID) REFERENCES usertbl (user ID)
- 외래키로 지정되어 있는 userID에 입력되는 새로운 값 STJ가 usertbl에 없는 값이기 때문에 무결성 오류 발생
    - buytbl테이블의 userid 컬럼은 usertbl테이블의 userID를 참조할때, usertbl테이블의 userID가 STJ인 데이터가 없으면 입력 안됨
    - 데이터 무결성이란? 두 테이블간 관계에 있어서 데이터의 정확성을 보장하는 제약 조건
    - 현업에서는 꼭 필요한 경우에만 사용하는 경우가 많음(비즈니스 로직이 다양하기 때문에 제약을 걸어놓을 경우, 예외적인 비즈니스 로직 처리가 어려움)

```python
# db연결을 활성화 해주는 함수 구현

def conn(d_name) : 
    import pymysql
    host_name = 'localhost'
    host_port = 3306
    username = 'root'
    password = 'toor'
    database_name = d_name
    db = pymysql.connect(
        host = host_name,      # MySQL Server Address
        port = host_port,      # MySQL Server Port
        user = username,       # password for MySQL username
        passwd = password,     # Database name
        db=database_name,
        charset='utf8'
    )
    return db
```

```python
db = conn('sqlDB')
```

**이번에는 usertbl에 userID가 STJ인 데이터를 넣어 준 후 다시 buytbl userID에 STJ 관련 데이터를 넣어줌**
- 에러 나지 않음

```python
cursor = db.cursor()
sql_query = "INSERT INTO usertbl VALUES('STJ', '서태지', 1975, '경기', '011', '00000000',171,'2015-5-20')"
cursor.execute(sql_query)
db.commit()
```

```python
SQL_QUERY = "INSERT INTO buytbl (userID, prodName, groupName, price, amount) VALUES('STJ', '운동화', null ,30, 3)"
cursor.execute(SQL_QUERY)
db.commit()
```

**이번에는 usertbl에 userID가 STJ 관련 데이터를 삭제 해 봄**
- 에러 나면 정상(buytlb에 해당 userID를 참조하는 데이터가 있기 때문)

```python
SQL_QUERY = "DELETE FROM usertbl WHERE userID='STJ'"
cursor.execute(SQL_QUERY)
db.commit()
```
