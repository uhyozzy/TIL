**1. 환경준비**
- 텐서플로 라이브러리 안에 있는 케라스 API에서 필요한 함수들을 불러옴
- 데이터를 다루는 데 필요한 라이브러리를 불러옴

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
```

**2. 데이터 준비**
- 준비된 수술 환자 데이터를 불러옴

```python
Data_set = np.loadtxt("./data/ThoraricSurgery3.csv", delimiter=",")
X=Data_set[:,0:16] # 환자의 진찰 기록을 X로 지정
y=Data_set[:,16] # 수술 1년 후 사망/생존 여부를 y로 지정
```

**3. 구조 결정**
- 딥러닝 모델의 구조를 결정

```python
model = Sequential()
model.add(Dense(30, input_dim=16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

**4. 모델 실행**
- 딥러닝 모델을 실행함

**1. model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']):**

- model.compile() 함수는 모델을 컴파일하는 단계입니다.
- loss='binary_crossentropy': 이진 분류 문제를 해결하는데 사용되는 손실 함수를 'binary_crossentropy'로 지정합니다. 이 함수는 이진 분류에서 자주 사용되는 손실 함수 중 하나로, 두 클래스 간의 크로스 엔트로피(Cross Entropy)를 계산합니다.
- optimizer='adam': 옵티마이저로 'adam'을 사용합니다. Adam은 경사 하강법의 변형으로, 학습 속도를 조정하여 빠르게 수렴하면서도 안정적으로 학습할 수 있는 최적화 알고리즘입니다.
- metrics=['accuracy']: 모델의 성능을 평가하는 지표로 'accuracy'를 사용합니다. 'accuracy'는 정확도를 나타내며, 모델의 예측이 실제 레이블과 얼마나 일치하는지를 나타냅니다.

**2. history=model.fit(X, y, epochs=30, batch_size=16):**

- model.fit() 함수는 모델을 주어진 데이터에 대해 훈련하는 단계입니다.
- X: 훈련 데이터로, 입력 특성들을 포함하는 NumPy 배열입니다.
- y: 훈련 데이터에 대한 레이블(타겟)로, 이진 분류 문제에서 0 또는 1의 값을 가지는 NumPy 배열입니다.
- epochs=30: 훈련 과정에서 전체 데이터셋을 몇 번 반복하여 학습할지를 나타냅니다. 이 경우 30번 반복합니다.
- batch_size=16: 한 번의 반복(에포크)에서 사용할 데이터 샘플의 개수를 나타냅니다. 이 값은 메모리 제약 등을 고려하여 적절히 선택해야 합니다.
- history: 모델의 훈련 과정에서 발생하는 손실과 성능 지표의 변화를 기록하는 객체입니다. 이를 이용하여 학습 과정을 시각화하거나 평가할 수 있습니다.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history=model.fit(X, y, epochs=30, batch_size=16)
```
